```python
from XGBoostRegressor import XGBoostRegressor
import sklearn.datasets
dataset = sklearn.datasets.load_boston()
data, target = dataset['data'], dataset['target']
model = XGBoostRegressor(num_iters=100, num_bins=10, lr=0.1, alpha=0.1, gamma=0.1, max_depth=6, anim=True, anim_y_lim=100)
model.fit(data, target)
result = model.transform(data)
```

```python
model.trees[0].plot(dataset['feature_names'])
```

```python
from Shap import Shap
shap = Shap(model.trees[0])
shap_res = shap.explain(data[0])
```
